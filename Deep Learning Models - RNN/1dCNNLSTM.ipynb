```python
import numpy as np
import pandas as pd
def create_windows(df, window_size, step_size=None, threshold=0.9):
    """
    Splits the dataset into overlapping or non-overlapping windows and generates corresponding labels.

    Parameters:
    - df: pandas.DataFrame
        The dataset with columns 'UHF', 'UV', 'UHL', and 'Annotation'.
    - window_size: int
        The size of each window.
    - step_size: int or None
        The step size for moving the window. Determines the amount of overlap.
        If None, defaults to window_size (no overlap).
        If step_size < window_size, there is overlap.
    - threshold: float
        The percentage threshold for assigning a label to a window (default is 0.9).

    Returns:
    - x_list: list
        List of windows for the 'UHF' column.
    - y_list: list
        List of windows for the 'UV' column.
    - z_list: list
        List of windows for the 'UHL' column.
    - train_labels: list
        List of labels for each window, determined by the ratio of annotations.
    """
    # Set default step_size for no overlap
    if step_size is None:
        step_size = window_size

    # Input validation
    if step_size <= 0:
        raise ValueError("Step size must be greater than 0.")
    if window_size > df.shape[0]:
        raise ValueError("Window size cannot be larger than the dataset.")

    x_list = []
    y_list = []
    z_list = []
    train_labels = []

    # Process dataset into windows
    for i in range(0, df.shape[0] - window_size + 1, step_size):
        # Extract window data
        xs = df['UHF'].values[i:i + window_size]
        ys = df['UV'].values[i:i + window_size]
        zs = df['UHL'].values[i:i + window_size]

        # Calculate annotation proportion for the window
        annotations = df['Annotation'][i:i + window_size]
        proportion = (annotations == 1).sum() / window_size

        # Assign label based on threshold
        train_labels.append(1 if proportion > threshold else 0)

        # Append window data to corresponding lists
        x_list.append(xs)
        y_list.append(ys)
        z_list.append(zs)

    return x_list, y_list, z_list, train_labels

def prepare_raw_data_for_dl(df, window_size, step_size=None, threshold=0.9):
    """
    Use the create_windows function to produce windows of raw sensor data,
    then arrange them into a 3D NumPy array (samples, timesteps, features)
    suitable for deep learning models.

    Parameters:
    - df: pandas.DataFrame
        The dataset with at least 'UHF', 'UV', 'UHL', 'Annotation' columns.
    - window_size: int
        The number of samples per window.
    - step_size: int or None
        Step size for windowing. If None, no overlap.
    - threshold: float
        Threshold for labeling a window as FOG.

    Returns:
    - X: np.ndarray
        The raw data in shape (num_samples, window_size, 3)
        where the third dimension corresponds to [UHF, UV, UHL].
    - y: np.ndarray
        Labels for each window (0 or 1).
    """
    x_list, y_list, z_list, train_labels = create_windows(df, window_size, step_size, threshold)

    # Convert to numpy arrays
    x_arr = np.array(x_list)  # shape: (num_samples, window_size)
    y_arr = np.array(y_list)
    z_arr = np.array(z_list)

    # Stack along the last axis to create a shape (samples, timesteps, features=3)
    # The features dimension is [UHF, UV, UHL]
    # x_arr is UHF, y_arr is UV, z_arr is UHL
    X = np.stack((x_arr, y_arr, z_arr), axis=2)  # shape: (num_samples, window_size, 3)

    # Convert labels to numpy array
    y = np.array(train_labels)

    return X, y

# Example usage (assuming df is your original DataFrame):
# window_size = 200 (for example)
# step_size = None (no overlap)
# X, y = prepare_raw_data_for_dl(df, window_size=200, threshold=0.9)
# Now X is a 3D array of raw data suitable for a CNN or LSTM.
```

```python
#nonstack
import numpy as np
from sklearn.model_selection import KFold
from sklearn.metrics import classification_report, precision_score, recall_score, f1_score, matthews_corrcoef
from tensorflow.keras.models import Sequential
from tensorflow.keras.layers import LSTM, Dense, Dropout, BatchNormalization
from tensorflow.keras.callbacks import Callback
from tensorflow.keras.optimizers import Adam
from sklearn.utils import class_weight

# Custom Callback to Calculate Metrics at the End of Each Epoch
class MetricsCallback(Callback):
    def __init__(self, X_val, y_val):
        super().__init__()
        self.X_val = X_val
        self.y_val = y_val

    def on_epoch_end(self, epoch, logs=None):
        y_pred = (self.model.predict(self.X_val) > 0.5).astype(int)
        precision = precision_score(self.y_val, y_pred)
        recall = recall_score(self.y_val, y_pred)
        mcc = matthews_corrcoef(self.y_val, y_pred)
        print(f"\nEpoch {epoch + 1}: Precision: {precision:.4f}, Recall: {recall:.4f}, MCC: {mcc:.4f}")

# Define N-fold Cross Validation Parameters
n_folds = 5
kf = KFold(n_splits=n_folds, shuffle=False)

# Placeholder for Metrics Aggregation
overall_metrics = {
    'precision': [],
    'recall': [],
    'f1_score': [],
    'mcc': []
}

# Perform Cross-Validation
fold = 1
for train_index, val_index in kf.split(X):
    print(f"\nFold {fold}/{n_folds}")
    
    # Split Data for the Current Fold
    X_train, X_val = X[train_index], X[val_index]
    y_train, y_val = y[train_index], y[val_index]

    # Compute Class Weights
    class_weights = class_weight.compute_class_weight('balanced', classes=np.unique(y_train), y=y_train)
    class_weights_dict = dict(enumerate(class_weights))



    # Build Model
    model = Sequential([
        LSTM(64, return_sequences=True, input_shape=(X_train.shape[1], X_train.shape[2])),
        BatchNormalization(),
        Dropout(0.3),
        LSTM(64),
        BatchNormalization(),
        Dropout(0.3),
        Dense(32, activation='relu'),
        Dense(1, activation='sigmoid')
    ])

    # Compile the Model
    model.compile(optimizer=Adam(), loss='binary_crossentropy', metrics=['accuracy'])

    # Metrics Callback
    metrics_callback = MetricsCallback(X_val, y_val)

    # Train the Model
    model.fit(
        X_train, y_train,
        validation_data=(X_val, y_val),
        epochs=20,
        batch_size=32,
        class_weight=class_weights_dict,
        callbacks=[metrics_callback],
        verbose=1
    )

    # Evaluate on Validation Set
    y_pred = (model.predict(X_val) > 0.5).astype(int)
    
    precision = precision_score(y_val, y_pred)
    recall = recall_score(y_val, y_pred)
    f1 = f1_score(y_val, y_pred)
    mcc = matthews_corrcoef(y_val, y_pred)

    # Print Classification Report for the Current Fold
    print("\nClassification Report for Fold {}:".format(fold))
    print(classification_report(y_val, y_pred))

    # Store Metrics
    overall_metrics['precision'].append(precision)
    overall_metrics['recall'].append(recall)
    overall_metrics['f1_score'].append(f1)
    overall_metrics['mcc'].append(mcc)

    fold += 1

# Print Overall Metrics
print("\nOverall Cross-Validation Metrics:")
print(f"Precision: {np.mean(overall_metrics['precision']):.4f} ± {np.std(overall_metrics['precision']):.4f}")
print(f"Recall: {np.mean(overall_metrics['recall']):.4f} ± {np.std(overall_metrics['recall']):.4f}")
print(f"F1 Score: {np.mean(overall_metrics['f1_score']):.4f} ± {np.std(overall_metrics['f1_score']):.4f}")
print(f"MCC: {np.mean(overall_metrics['mcc']):.4f} ± {np.std(overall_metrics['mcc']):.4f}")
```

```python
from tensorflow.keras.models import Model
from tensorflow.keras.layers import (
    Input, Conv1D, BatchNormalization, Activation, Dropout, Flatten, Dense,
    Add, concatenate, Lambda, multiply
)
from tensorflow.keras.callbacks import EarlyStopping
from tensorflow.keras.regularizers import l2
import tensorflow.keras.backend as K

# Define a TCN block
def tcn_block(inputs, num_filters, kernel_size, dilation_rate, dropout_rate, l2_reg):
    x = Conv1D(filters=num_filters, kernel_size=kernel_size, 
               dilation_rate=dilation_rate, padding="causal", 
               kernel_regularizer=l2(l2_reg))(inputs)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Second convolution in the block
    x = Conv1D(filters=num_filters, kernel_size=kernel_size, 
               dilation_rate=dilation_rate, padding="causal", 
               kernel_regularizer=l2(l2_reg))(x)
    x = BatchNormalization()(x)
    x = Activation('relu')(x)
    x = Dropout(dropout_rate)(x)

    # Residual connection
    res = Conv1D(filters=num_filters, kernel_size=1, padding="same", 
                 kernel_regularizer=l2(l2_reg))(inputs)
    x = Add()([x, res])
    return x

# Define an attention mechanism for the temporal data
def attention_block(inputs):
    attention = Dense(1, activation='sigmoid')(inputs)
    attention = Lambda(lambda x: K.expand_dims(x, axis=-1))(attention)  # Expand last dimension
    return multiply([inputs, attention])

# Input for raw accelerometer data
raw_data_input = Input(shape=(200, 3), name="raw_data_input")
x = tcn_block(raw_data_input, num_filters=64, kernel_size=3, dilation_rate=1, dropout_rate=0.3, l2_reg=0.01)
x = tcn_block(x, num_filters=64, kernel_size=3, dilation_rate=2, dropout_rate=0.3, l2_reg=0.01)
x = attention_block(x)  # Apply attention mechanism
x = Flatten()(x)
raw_data_output = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(x)

# Input for feature-engineered data
feature_input = Input(shape=(149,), name="feature_input")
y = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(feature_input)
y = BatchNormalization()(y)
y = Dropout(0.3)(y)
feature_output = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(y)

# Combine the two heads
combined = concatenate([raw_data_output, feature_output])
z = Dense(128, activation='relu', kernel_regularizer=l2(0.01))(combined)
z = Dropout(0.3)(z)
z = Dense(64, activation='relu', kernel_regularizer=l2(0.01))(z)
z = Dropout(0.3)(z)
final_output = Dense(1, activation='sigmoid', name="final_output")(z)

# Build the model
model = Model(inputs=[raw_data_input, feature_input], outputs=final_output)

# Compile the model
model.compile(optimizer='adam', loss='binary_crossentropy', metrics=['accuracy'])

# Print the model summary
model.summary()

# Early stopping
early_stopping = EarlyStopping(monitor='val_loss', patience=5, restore_best_weights=True)

# Train the model
history = model.fit(
    [X, features], y,  # Multi-input: raw data and features
    validation_split=0.2,
    epochs=20,
    batch_size=32,
    class_weight=class_weights_dict,
    callbacks=[early_stopping]
)

# Evaluate the model
y_pred = (model.predict([X_test, features_test]) > 0.5).astype(int)
print("\nFinal Classification Report:")
print(classification_report(y_test, y_pred))
```
