```python
def create_windows(df, window_size, step_size=None, threshold=0.9):
    """
    Splits the dataset into overlapping or non-overlapping windows and generates corresponding labels.

    Parameters:
    - df: pandas.DataFrame
        The dataset with columns 'UHF', 'UV', 'UHL', and 'Annotation'.
    - window_size: int
        The size of each window.
    - step_size: int or None
        The step size for moving the window. Determines the amount of overlap.
        If None, defaults to window_size (no overlap).
        If step_size < window_size, there is overlap.
    - threshold: float
        The percentage threshold for assigning a label to a window (default is 0.9).

    Returns:
    - x_list: list
        List of windows for the 'UHF' column.
    - y_list: list
        List of windows for the 'UV' column.
    - z_list: list
        List of windows for the 'UHL' column.
    - train_labels: list
        List of labels for each window, determined by the ratio of annotations.
    """
    # Set default step_size for no overlap
    if step_size is None:
        step_size = window_size

    # Input validation
    if step_size <= 0:
        raise ValueError("Step size must be greater than 0.")
    if window_size > df.shape[0]:
        raise ValueError("Window size cannot be larger than the dataset.")

    x_list = []
    y_list = []
    z_list = []
    train_labels = []

    # Process dataset into windows
    for i in range(0, df.shape[0] - window_size + 1, step_size):
        # Extract window data
        xs = df['UHF'].values[i:i + window_size]
        ys = df['UV'].values[i:i + window_size]
        zs = df['UHL'].values[i:i + window_size]

        # Calculate annotation proportion for the window
        annotations = df['Annotation'][i:i + window_size]
        proportion = (annotations == 1).sum() / window_size

        # Assign label based on threshold
        train_labels.append(1 if proportion > threshold else 0)

        # Append window data to corresponding lists
        x_list.append(xs)
        y_list.append(ys)
        z_list.append(zs)

    return x_list, y_list, z_list, train_labels
```

```python
from scipy import stats, signal
import numpy as np
from scipy.signal import welch
from scipy import integrate
import pandas as pd
from scipy.signal import find_peaks, butter, filtfilt, welch


def feature_engineering(x_list, y_list, z_list, sampling_rate=64):
    # Convert lists to arrays and handle NaN
    x_arr, y_arr, z_arr = np.nan_to_num(np.array(x_list)), np.nan_to_num(np.array(y_list)), np.nan_to_num(np.array(z_list))
    window_size = len(x_arr)
    features = []
    
    # Filtering functions (original from Function 1)
    def butter_filter(data, cutoff, fs, order, filter_type):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype=filter_type, analog=False)
        return filtfilt(b, a, data)

    # Apply low-pass and high-pass filters
    x_filt = butter_filter(x_arr, cutoff=15.0, fs=sampling_rate, order=2, filter_type='low')
    y_filt = butter_filter(y_arr, cutoff=15.0, fs=sampling_rate, order=2, filter_type='low')
    z_filt = butter_filter(z_arr, cutoff=15.0, fs=sampling_rate, order=2, filter_type='low')

    x_filt = butter_filter(x_filt, cutoff=0.2, fs=sampling_rate, order=3, filter_type='high')
    y_filt = butter_filter(y_filt, cutoff=0.2, fs=sampling_rate, order=3, filter_type='high')
    z_filt = butter_filter(z_filt, cutoff=0.2, fs=sampling_rate, order=3, filter_type='high')

    # Time-domain features (original)
    stats_dict = {}
    for axis, arr in zip(['x', 'y', 'z'], [x_filt, y_filt, z_filt]):
        axis_mean = np.mean(arr)
        axis_std = np.std(arr)
        # Data-driven prominence for peak detection
        peak_prominence = 0.5 * axis_std if axis_std > 0 else 0.01
        peaks, _ = find_peaks(arr, prominence=peak_prominence)
        
        # Zero-crossing rate (ZCR)
        zcr = np.sum(np.diff(np.sign(arr)) != 0) / window_size
        
        stats_dict[axis] = {
            'mean': axis_mean,
            'std': axis_std,
            'median': np.median(arr),
            'min': np.min(arr),
            'max': np.max(arr),
            'mad': np.median(np.abs(arr - np.median(arr))),
            'aad': np.mean(np.abs(arr - axis_mean)),
            'peaks': len(peaks),
            'zcr': zcr
        }

    for axis in ['x', 'y', 'z']:
        s = stats_dict[axis]
        features += [s['mean'], s['std'], s['mad'], s['aad'], s['min'], s['max']]
        features.append(s['max'] - s['min'])  # range
        features.append(s['peaks'])            # peak count
        features.append(s['zcr'])              # zero-crossing rate
    
    # Resultant and SMA (original)
    resultant = np.sqrt(x_filt**2 + y_filt**2 + z_filt**2)
    features.append(np.mean(resultant))  # Avg resultant
    sma = (np.sum(np.abs(x_filt)) + np.sum(np.abs(y_filt)) + np.sum(np.abs(z_filt))) / window_size
    features.append(sma)  # Signal Magnitude Area
    
    def compute_octave_freeze_index(arr, fs=sampling_rate, window_len=window_size, loco_band=(0.5, 3), freeze_band=(3, 8), power_threshold=2**11.5):
        fft_vals = np.fft.fft(arr, n=window_len)
        power_spectrum = np.abs(fft_vals) ** 2 / window_len
        freqs = np.fft.fftfreq(window_len, d=1/fs)

        # Compute band indices
        f_res = fs / window_len
        loco_start = int(round(loco_band[0] / f_res))
        loco_end = int(round(loco_band[1] / f_res))
        freeze_start = int(round(freeze_band[0] / f_res))
        freeze_end = int(round(freeze_band[1] / f_res))

        # Calculate band energies
        loco_energy = np.sum(power_spectrum[loco_start:loco_end])
        freeze_energy = np.sum(power_spectrum[freeze_start:freeze_end])

        # Apply energy threshold
        total_energy = loco_energy + freeze_energy
        if total_energy < power_threshold:
            freeze_index = 0
        else:
            freeze_index = freeze_energy / loco_energy if loco_energy > 0 else 0

        return freeze_index

    
    def otherfreeze_index(data, sf, band1, band2, win_sec=None, relative=False):
        """
        The power in the “freeze” band (band1) divided by the power in the locomotor band (band2).
        
        Parameters:
        ----------
        data : 1d-array
            Input signal in the time-domain.
        sf : float
            Sampling frequency of the data.
        band1 : list
            Lower and upper frequencies of the freeze band (e.g., [3, 8]).
        band2 : list
            Lower and upper frequencies of the locomotor band (e.g., [0.5, 3]).
        win_sec : float, optional
            Length of each window in seconds. If None, defaults to (1 / min(band)) * 2.
        relative : boolean, optional
            If True, computes relative power.
        
        Returns:
        -------
        fi : float
            Ratio of power in band1 to band2.
        """
        if not isinstance(data, np.ndarray):
            raise ValueError("Input 'data' must be a NumPy array.")
        if len(band1) != 2 or len(band2) != 2:
            raise ValueError("Bands must be specified as [low, high].")
    
        # Compute freeze index as the ratio of band powers
        fi = _bandpower(data, sf, band1, win_sec, relative) / _bandpower(data, sf, band2, win_sec, relative)
        return fi
    
    
    def _bandpower(data, sf, band, window_sec=None, relative=False):
        """
        Compute the average power of the signal in a specific frequency band.
    
        Parameters:
        ----------
        data : 1d-array
            Input signal in the time-domain.
        sf : float
            Sampling frequency of the data.
        band : list
            Lower and upper frequencies of the band of interest.
        window_sec : float, optional
            Length of each window in seconds. If None, defaults to (1 / min(band)) * 2.
        relative : boolean, optional
            If True, computes relative power.
    
        Returns:
        -------
        bp : float
            Absolute or relative band power.
        """
        band = np.asarray(band)
        low, high = band
    
        if low >= high:
            raise ValueError("The lower band edge must be less than the upper band edge.")
        if np.any(band < 0):
            raise ValueError("Frequencies in the band must be non-negative.")
    
        # Define window length
        if window_sec is not None:
            nperseg = int(window_sec * sf)
        else:
            nperseg = int((2 / low) * sf)
    
        # Compute the modified periodogram (Welch's method)
        freqs, psd = welch(data, sf, nperseg=nperseg)
    
        # Frequency resolution
        freq_res = freqs[1] - freqs[0]
    
        # Find indices for the band of interest
        idx_band = np.logical_and(freqs >= low, freqs <= high)
    
        # Compute the band power using the trapezoidal rule for integration
        bp = np.trapz(psd[idx_band], dx=freq_res)
    
        if relative:
            # Compute total power for normalization
            bp /= np.trapz(psd, dx=freq_res)
        return bp
    
    


    ########################################
    # Original FFT-based features
    ########################################
    def compute_fft_features_original(arr, fs=sampling_rate):
        fft_vals = np.fft.fft(arr)
        freqs = np.fft.fftfreq(window_size, d=1.0/fs)
        
        # Positive half spectrum
        half_mask = freqs > 0
        freqs = freqs[half_mask]
        fft_mag = np.abs(fft_vals[half_mask])
        power_spectrum = fft_mag**2
        
        # Band energies
        low_band = (freqs >= 0.5) & (freqs <= 3)
        fog_band = (freqs > 3) & (freqs <= 8)
        
        low_energy = np.sum(power_spectrum[low_band])
        fog_energy = np.sum(power_spectrum[fog_band])
        total_energy = np.sum(power_spectrum)
        
        # Freeze Index
        freeze_index = fog_energy / (low_energy + 1e-12)
        
        # Spectral Entropy
        if total_energy > 0:
            p = power_spectrum / total_energy
            spectral_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            spectral_entropy = 0.0
        
        # Spectral Centroid
        spectral_centroid = np.sum(freqs * fft_mag) / np.sum(fft_mag) if np.sum(fft_mag) > 0 else 0.0

        return [low_energy, fog_energy, freeze_index, spectral_entropy, spectral_centroid]

    ########################################
    # Additional FFT Features (from Function 2)
    ########################################
    def compute_fft_features_additional(arr, fs=sampling_rate):
        fft_vals = np.fft.fft(arr)
        freqs = np.fft.fftfreq(window_size, d=1.0/fs)
        half_mask = freqs > 0
        freqs = freqs[half_mask]
        fft_mag = np.abs(fft_vals[half_mask])
        power_spectrum = fft_mag**2
        
        # Basic FFT stats
        fft_mean = np.mean(fft_mag)
        fft_std = np.std(fft_mag)
        fft_median = np.median(fft_mag)
        fft_range = (np.max(fft_mag) - np.min(fft_mag)) if len(fft_mag) > 0 else 0.0

        # Dominant frequency and amplitude
        if len(fft_mag) > 0:
            max_idx = np.argmax(fft_mag)
            dominant_freq = freqs[max_idx]
            dominant_amp = fft_mag[max_idx]
        else:
            dominant_freq = 0.0
            dominant_amp = 0.0
        
        # Additional bands: Mid (3–10 Hz), High (>10 Hz)
        low_band = (freqs >= 0) & (freqs <= 3)
        mid_band = (freqs > 3) & (freqs <= 10)
        high_band = (freqs > 10)
        
        low_energy = np.sum(power_spectrum[low_band])
        mid_energy = np.sum(power_spectrum[mid_band])
        high_energy = np.sum(power_spectrum[high_band])
        total_energy = np.sum(power_spectrum)
        
        # Spectral Entropy and Centroid (for additional features)
        if total_energy > 0:
            p = power_spectrum / total_energy
            spectral_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            spectral_entropy = 0.0
        
        if np.sum(fft_mag) > 0:
            spectral_centroid = np.sum(freqs * fft_mag) / np.sum(fft_mag)
        else:
            spectral_centroid = 0.0
        
        # Spectral Flatness
        nonzero_mag = fft_mag[fft_mag > 0]
        if len(nonzero_mag) > 0:
            geom_mean = np.exp(np.mean(np.log(nonzero_mag)))
            spectral_flatness = geom_mean / (fft_mean + 1e-12)
        else:
            spectral_flatness = 0.0
        
        # Spectral Spread
        if np.sum(fft_mag) > 0:
            spectral_spread = np.sqrt(np.sum(((freqs - spectral_centroid)**2) * fft_mag) / np.sum(fft_mag))
        else:
            spectral_spread = 0.0
        
        # Top N FFT coefficients
        N = 5
        sorted_fft = np.sort(fft_mag)
        if len(sorted_fft) >= N:
            top_n_coeffs = sorted_fft[-N:]
        else:
            top_n_coeffs = np.pad(sorted_fft, (N-len(sorted_fft), 0), 'constant')

        return [
            fft_mean, fft_std, fft_median, fft_range,
            dominant_freq, dominant_amp,
            low_energy, mid_energy, high_energy,
            spectral_entropy, spectral_centroid,
            spectral_flatness, spectral_spread
        ] + list(top_n_coeffs)

    def compute_fft_features_combined(arr, fs=sampling_rate):
        original_feats = compute_fft_features_original(arr, fs)
        additional_feats = compute_fft_features_additional(arr, fs)
        return original_feats + additional_feats

    # Compute combined FFT features for each axis
    for arr in [x_filt, y_filt, z_filt]:
        features += compute_fft_features_combined(arr, fs=sampling_rate)
        # Compute Freeze Index for x, y, z


    ########################################
    # PSD-based features (original)
    ########################################
    def compute_psd_features_original(arr, fs=sampling_rate):
        f_welch, Pxx = welch(arr, fs=fs, nperseg=window_size)
        
        low_idx = (f_welch >= 0.5) & (f_welch <= 3)
        fog_idx = (f_welch > 3) & (f_welch <= 8)
        
        low_power = np.sum(Pxx[low_idx])
        fog_power = np.sum(Pxx[fog_idx])
        total_power = np.sum(Pxx)
        
        freeze_index = fog_power / (low_power + 1e-12)
        low_rel = low_power / total_power if total_power > 0 else 0.0
        fog_rel = fog_power / total_power if total_power > 0 else 0.0
        
        psd_centroid = np.sum(f_welch * Pxx) / total_power if total_power > 0 else 0.0
        
        return [low_rel, fog_rel, freeze_index, psd_centroid]

    ########################################
    # Additional PSD Features (from Function 2)
    ########################################
    def compute_psd_features_additional(arr, fs=sampling_rate):
        f_welch, Pxx = welch(arr, fs=fs, nperseg=window_size)
        total_power = np.sum(Pxx)

        low_idx = (f_welch >= 0) & (f_welch <= 3)
        mid_idx = (f_welch > 3) & (f_welch <= 10)
        high_idx = (f_welch > 10)

        low_power = np.sum(Pxx[low_idx])
        mid_power = np.sum(Pxx[mid_idx])
        high_power = np.sum(Pxx[high_idx])
        
        if total_power > 0:
            low_rel = low_power / total_power
            mid_rel = mid_power / total_power
            high_rel = high_power / total_power
        else:
            low_rel = mid_rel = high_rel = 0.0

        if len(Pxx) > 0:
            max_idx = np.argmax(Pxx)
            peak_freq = f_welch[max_idx]
            peak_power = Pxx[max_idx]
        else:
            peak_freq = 0.0
            peak_power = 0.0
        
        if total_power > 0:
            p = Pxx / total_power
            psd_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            psd_entropy = 0.0
        
        low_high_ratio = low_power / high_power if high_power > 0 else 0.0
        low_mid_ratio = low_power / mid_power if mid_power > 0 else 0.0

        if total_power > 0:
            psd_centroid = np.sum(f_welch * Pxx) / total_power
            psd_spread = np.sqrt(np.sum((f_welch - psd_centroid)**2 * Pxx) / total_power)
        else:
            psd_centroid = 0.0
            psd_spread = 0.0
        
        nonzero_pxx = Pxx[Pxx > 0]
        if len(nonzero_pxx) > 0:
            geom_mean = np.exp(np.mean(np.log(nonzero_pxx)))
            psd_flatness = geom_mean / (np.mean(Pxx) + 1e-12)
        else:
            psd_flatness = 0.0

        return [
            total_power,
            low_rel, mid_rel, high_rel,
            peak_freq, peak_power,
            psd_entropy,
            low_high_ratio, low_mid_ratio,
            psd_centroid, psd_spread, psd_flatness
        ]

    def compute_psd_features_combined(arr, fs=sampling_rate):
        original_feats = compute_psd_features_original(arr, fs)
        additional_feats = compute_psd_features_additional(arr, fs)
        return original_feats + additional_feats

    # Compute combined PSD features for each axis
    for arr in [x_filt, y_filt, z_filt]:
        features += compute_psd_features_combined(arr, fs=sampling_rate)
        
    for arr in [x_filt, y_filt, z_filt]:
        freeze_index = compute_octave_freeze_index(arr)
        features.append(freeze_index)
        
    win_sec = float(window_size/sampling_rate)
    for arr in [x_filt, y_filt, z_filt]:
        freeze_index = otherfreeze_index(arr, sampling_rate, [3, 8], [0.5, 3], win_sec)                                                              
        features.append(freeze_index)

    # Construct column names
    axes = ['x', 'y', 'z']
    time_features = ['mean', 'std', 'mad', 'aad', 'min', 'max', 'range', 'peaks', 'zcr']
    # 3 axes * 9 features each = 27, plus 2 global (mean_resultant, sma) = 29
    columns = []
    for axis in axes:
        for feat in time_features:
            columns.append(f"{axis}_{feat}")
    columns += ['mean_resultant', 'sma']

    # Original FFT-based features (5 per axis)
    fft_feats_original = ['fft_low_energy', 'fft_fog_energy', 'fft_freeze_index', 'fft_spectral_entropy', 'fft_spectral_centroid']

    # Additional FFT features
    fft_additional_feats = [
        'fft_mean', 'fft_std', 'fft_median', 'fft_range',
        'fft_dominant_freq', 'fft_dominant_amp',
        'fft_low_energy_ext', 'fft_mid_energy', 'fft_high_energy',
        'fft_spectral_entropy_ext', 'fft_spectral_centroid_ext',
        'fft_spectral_flatness', 'fft_spectral_spread'
    ]
    for i in range(1, 6):
        fft_additional_feats.append(f"fft_top_coeff_{i}")

    # Columns for FFT (original + additional)
    for axis in axes:
        for feat in fft_feats_original:
            columns.append(f"{axis}_{feat}")
        for feat in fft_additional_feats:
            columns.append(f"{axis}_{feat}")

    # Original PSD features (4 per axis)
    psd_feats_original = ['psd_low_rel', 'psd_fog_rel', 'psd_freeze_index', 'psd_centroid']

    # Additional PSD features
    psd_additional_feats = [
        'psd_total_power',
        'psd_low_rel_ext', 'psd_mid_rel', 'psd_high_rel',
        'psd_peak_freq', 'psd_peak_power',
        'psd_entropy',
        'psd_low_high_ratio', 'psd_low_mid_ratio',
        'psd_centroid_ext', 'psd_spread', 'psd_flatness'
    ]

    # Columns for PSD (original + additional)
    for axis in axes:
        for feat in psd_feats_original:
            columns.append(f"{axis}_{feat}")
        for feat in psd_additional_feats:
            columns.append(f"{axis}_{feat}")

    columns.append("Octavex")
    columns.append("Octavey")
    columns.append("Octavez")
    
    columns.append("otherx")
    columns.append("othery")
    columns.append("otherz")

    # Create a DataFrame from the features
    df = pd.DataFrame([features], columns=columns)

    return df
```

```python
#import numpy as np
from sklearn.metrics import matthews_corrcoef
import numpy as np
from sklearn.metrics import confusion_matrix
import numpy as np
from sklearn.metrics import confusion_matrix

def sens(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    sensitivity = tp / (tp + fn)
    return sensitivity

 
def spec(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    specificity = tn / (tn + fp)
    return specificity   

def f1(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    f1_score = 2 * ((sensitivity * specificity) / (sensitivity + specificity))
    return f1_score  

def mcc(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    mcc = (tp * tn - fp * fn) / np.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
    return mcc  
def acc(y_true,y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    return accuracy  

def classification_metrics(y_true, y_pred):
    cm = confusion_matrix(y_true, y_pred)
    tn, fp, fn, tp = cm.ravel()
    
    sensitivity = tp / (tp + fn)
    specificity = tn / (tn + fp)
    f1_score = 2 * ((sensitivity * specificity) / (sensitivity + specificity))
    accuracy = (tp + tn) / (tp + tn + fp + fn)
    mcc = (tp * tn - fp * fn) / np.sqrt((tp + fp) * (tp + fn) * (tn + fp) * (tn + fn))
    
    print("Sensitivity: {:.2f}".format(sensitivity))
    print("Specificity: {:.2f}".format(specificity))
    print("F1 score: {:.2f}".format(f1_score))
    print("Accuracy: {:.2f}".format(accuracy))
    print("MCC score: {:.2f}".format(mcc))
```

```python
import numpy as np
import matplotlib.pyplot as plt

def plot_feature_with_fog_events(Xn, y_true, y_pred, feature_index, min_length=3):
    """
    Plot a specified feature (by column index) over window number and overlay FOG events (true and predicted).

    Parameters:
    - Xn: np.ndarray
        2D array where each row is a window and each column is a feature.
        Shape: (num_windows, num_features)
    - y_true: np.ndarray
        True labels per window (0 or 1), shape (num_windows,).
    - y_pred: np.ndarray
        Predicted labels per window (0 or 1), shape (num_windows,).
    - feature_index: int
        The index of the column in Xn that you want to plot.
    - min_length: int
        Minimum number of continuous '1's to be considered a FOG event.
    """

    num_windows, num_features = Xn.shape
    if feature_index < 0 or feature_index >= num_features:
        raise ValueError(f"feature_index {feature_index} is out of range. Must be between 0 and {num_features-1}.")

    # Extract the feature values
    feature_values = Xn[:, feature_index]
    x = np.arange(num_windows)

    # Function to find continuous segments of length >= min_length where arr == 1
    def find_fog_segments(arr):
        segments = []
        start = None
        for i in range(num_windows):
            if arr[i] == 1 and start is None:
                start = i
            if arr[i] == 0 and start is not None:
                # end of a segment
                end = i - 1
                length = end - start + 1
                if length >= min_length:
                    segments.append((start, end))
                start = None
        # Check if ended at the end of the array
        if start is not None:
            end = num_windows - 1
            length = end - start + 1
            if length >= min_length:
                segments.append((start, end))
        return segments

    true_segments = find_fog_segments(y_true)
    pred_segments = find_fog_segments(y_pred)

    # Plotting
    plt.figure(figsize=(12, 6))
    plt.plot(x, feature_values, label=f'Feature {feature_index}', color='black')

    # Keep track of which labels have been added for the legend
    true_label_added = False
    pred_label_added = False

    # Overlay true FOG events
    for (s, e) in true_segments:
        if not true_label_added:
            plt.axvspan(s, e, color='red', alpha=0.2, label='True FOG Event')
            true_label_added = True
        else:
            plt.axvspan(s, e, color='red', alpha=0.2)

    # Overlay predicted FOG events
    for (s, e) in pred_segments:
        if not pred_label_added:
            plt.axvspan(s, e, color='blue', alpha=0.2, label='Predicted FOG Event')
            pred_label_added = True
        else:
            plt.axvspan(s, e, color='blue', alpha=0.2)

    plt.xlabel('Window Number')
    plt.ylabel(f'Feature {feature_index}')
    plt.title(f'Feature {feature_index} with True and Predicted FOG Events')
    plt.legend()
    plt.tight_layout()
    plt.show()

# Example usage:
# Xn = np.random.rand(200, 150)  # 200 windows, 150 features
# y_true = np.zeros(200); y_true[50:55] = 1  # short FOG event
# y_pred = np.zeros(200); y_pred[100:103] = 1
# plot_feature_with_fog_events(Xn, y_true, y_pred, feature_index=2, min_length=3)
```

```python
from scipy import stats, signal
import numpy as np
from scipy.signal import find_peaks, butter, filtfilt, welch

def feature_engineering(x_list, y_list, z_list, sampling_rate=66.7):
    # Convert lists to arrays and handle NaN
    x_arr, y_arr, z_arr = np.nan_to_num(np.array(x_list)), np.nan_to_num(np.array(y_list)), np.nan_to_num(np.array(z_list))
    window_size = len(x_arr)
    features = []
    
    # Low-pass filter to remove high-frequency noise
    def low_pass_filter(data, cutoff=15.0, fs=sampling_rate, order=4):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype='low', analog=False)
        return filtfilt(b, a, data)
    
    x_filt = low_pass_filter(x_arr)
    y_filt = low_pass_filter(y_arr)
    z_filt = low_pass_filter(z_arr)

    # Time-domain features
    stats_dict = {}
    for axis, arr in zip(['x', 'y', 'z'], [x_filt, y_filt, z_filt]):
        axis_mean = np.mean(arr)
        axis_std = np.std(arr)
        
        # Data-driven prominence for peak detection
        peak_prominence = 0.5 * axis_std if axis_std > 0 else 0.01
        peaks, _ = find_peaks(arr, prominence=peak_prominence)
        
        # Zero-crossing rate (ZCR)
        zcr = np.sum(np.diff(np.sign(arr)) != 0) / window_size
        
        stats_dict[axis] = {
            'mean': axis_mean,
            'std': axis_std,
            'median': np.median(arr),
            'min': np.min(arr),
            'max': np.max(arr),
            'mad': np.median(np.abs(arr - np.median(arr))),
            'aad': np.mean(np.abs(arr - axis_mean)),
            'peaks': len(peaks),
            'zcr': zcr
        }

    for axis in ['x', 'y', 'z']:
        s = stats_dict[axis]
        features += [s['mean'], s['std'], s['mad'], s['aad'], s['min'], s['max']]
        features.append(s['max'] - s['min'])  # range
        features.append(s['peaks'])          # peak count
        features.append(s['zcr'])            # zero-crossing rate
    
    # Resultant and SMA
    resultant = np.sqrt(x_filt**2 + y_filt**2 + z_filt**2)
    features.append(np.mean(resultant))  # Avg resultant
    sma = (np.sum(np.abs(x_filt)) + np.sum(np.abs(y_filt)) + np.sum(np.abs(z_filt))) / window_size
    features.append(sma)  # Signal Magnitude Area

    ########################################
    # Frequency-domain features (FFT-based)
    ########################################
    def compute_fft_features(arr, fs=sampling_rate):
        fft_vals = np.fft.fft(arr)
        freqs = np.fft.fftfreq(window_size, d=1.0/fs)
        
        # Positive half spectrum
        half_mask = freqs > 0
        freqs = freqs[half_mask]
        fft_mag = np.abs(fft_vals[half_mask])
        power_spectrum = fft_mag**2
        
        # Band energies
        low_band = (freqs >= 0) & (freqs <= 3)
        fog_band = (freqs >= 6) & (freqs <= 8)
        
        low_energy = np.sum(power_spectrum[low_band])
        fog_energy = np.sum(power_spectrum[fog_band])
        total_energy = np.sum(power_spectrum)
        
        # Energy ratios
        low_rel = low_energy / total_energy if total_energy > 0 else 0.0
        fog_rel = fog_energy / total_energy if total_energy > 0 else 0.0
        
        # Dominant frequency and amplitude
        dominant_freq = freqs[np.argmax(fft_mag)]
        is_fog_freq = 1 if 6 <= dominant_freq <= 8 else 0
        
        # Spectral Entropy
        if total_energy > 0:
            p = power_spectrum / total_energy
            spectral_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            spectral_entropy = 0.0
        
        return [
            low_rel, fog_rel, dominant_freq, is_fog_freq, spectral_entropy
        ]
    
    for arr in [x_filt, y_filt, z_filt]:
        features += compute_fft_features(arr, fs=sampling_rate)

    ########################################
    # Power Spectral Density (PSD-based)
    ########################################
    def compute_psd_features(arr, fs=sampling_rate):
        f_welch, Pxx = welch(arr, fs=fs, nperseg=window_size)
        
        # Frequency band indices
        low_idx = (f_welch >= 0) & (f_welch <= 3)
        fog_idx = (f_welch >= 6) & (f_welch <= 8)
        
        # Band powers
        low_power = np.sum(Pxx[low_idx])
        fog_power = np.sum(Pxx[fog_idx])
        total_power = np.sum(Pxx)
        
        # Power ratios
        low_rel = low_power / total_power if total_power > 0 else 0.0
        fog_rel = fog_power / total_power if total_power > 0 else 0.0
        
        # Peak frequency
        peak_freq = f_welch[np.argmax(Pxx)]
        
        return [low_rel, fog_rel, peak_freq]

    for arr in [x_filt, y_filt, z_filt]:
        features += compute_psd_features(arr, fs=sampling_rate)

    return np.array(features)
```

```python
from scipy import stats, signal
import numpy as np
from scipy.signal import find_peaks, butter, filtfilt, welch

def feature_engineering(x_list, y_list, z_list, sampling_rate=66.7):
    # Convert lists to arrays and handle NaN
    x_arr, y_arr, z_arr = np.nan_to_num(np.array(x_list)), np.nan_to_num(np.array(y_list)), np.nan_to_num(np.array(z_list))
    window_size = len(x_arr)
    features = []
    
    # Optional: Apply a low-pass filter to smoothen signals (if desired)
    def low_pass_filter(data, cutoff=6.0, fs=sampling_rate, order=4):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype='low', analog=False)
        return filtfilt(b, a, data)
    
    x_filt = low_pass_filter(x_arr)
    y_filt = low_pass_filter(y_arr)
    z_filt = low_pass_filter(z_arr)

    # Compute basic time-domain stats and peaks
    stats_dict = {}
    for axis, arr in zip(['x', 'y', 'z'], [x_filt, y_filt, z_filt]):
        axis_mean = np.mean(arr)
        axis_std = np.std(arr)
        
        # Data-driven prominence for peak detection
        peak_prominence = 0.5 * axis_std if axis_std > 0 else 0.01
        peaks, _ = find_peaks(arr, prominence=peak_prominence)
        
        stats_dict[axis] = {
            'mean': axis_mean,
            'std': axis_std,
            'median': np.median(arr),
            'min': np.min(arr),
            'max': np.max(arr),
            'mad': np.median(np.abs(arr - np.median(arr))),
            'aad': np.mean(np.abs(arr - axis_mean)),
            'peaks': len(peaks)
        }

    # Time-domain features
    for axis in ['x', 'y', 'z']:
        s = stats_dict[axis]
        features += [s['mean'], s['std'], s['mad'], s['aad'], s['min'], s['max']]
        features.append(s['max'] - s['min'])  # range
        features.append(s['peaks'])
    
    # Resultant and SMA
    resultant = np.sqrt(x_arr**2 + y_arr**2 + z_arr**2)
    features.append(np.mean(resultant))  # Avg resultant
    sma = (np.sum(np.abs(x_arr)) + np.sum(np.abs(y_arr)) + np.sum(np.abs(z_arr))) / window_size
    features.append(sma)  # Signal Magnitude Area

    ########################################
    # Frequency-domain features (FFT-based) #
    ########################################
    # Helper functions for frequency-based features
    def compute_fft_features(arr, fs=sampling_rate):
        # Compute FFT
        fft_vals = np.fft.fft(arr)
        freqs = np.fft.fftfreq(window_size, d=1.0/fs)
        
        # Consider only the positive half of the spectrum
        half_mask = freqs > 0
        freqs = freqs[half_mask]
        fft_mag = np.abs(fft_vals[half_mask])
        
        # Basic FFT stats (already included previously)
        fft_mean = np.mean(fft_mag)
        fft_std = np.std(fft_mag)
        fft_median = np.median(fft_mag)
        fft_range = np.max(fft_mag) - np.min(fft_mag)
        
        # Dominant frequency and amplitude
        max_idx = np.argmax(fft_mag)
        dominant_freq = freqs[max_idx]
        dominant_amp = fft_mag[max_idx]
        
        # Energy in predefined bands (using FFT magnitude squared as proxy for power)
        # Define bands: Low (0–3 Hz), Mid (3–10 Hz), High (>10 Hz)
        power_spectrum = fft_mag**2
        low_band = (freqs >= 0) & (freqs <= 3)
        mid_band = (freqs > 3) & (freqs <= 10)
        high_band = (freqs > 10)
        
        low_energy = np.sum(power_spectrum[low_band])
        mid_energy = np.sum(power_spectrum[mid_band])
        high_energy = np.sum(power_spectrum[high_band])
        
        total_energy = np.sum(power_spectrum)
        
        # Spectral Entropy (from FFT)
        # Normalize power spectrum
        if total_energy > 0:
            p = power_spectrum / total_energy
            spectral_entropy = -np.sum(p * np.log2(p + 1e-12))  # add small constant to avoid log(0)
        else:
            spectral_entropy = 0.0
        
        # Spectral Centroid (using FFT)
        # centroid = sum(f * A(f)) / sum(A(f))
        if np.sum(fft_mag) > 0:
            spectral_centroid = np.sum(freqs * fft_mag) / np.sum(fft_mag)
        else:
            spectral_centroid = 0.0
        
        # Top N FFT coefficients (let’s say N=5)
        N = 5
        top_n_coeffs = np.sort(fft_mag)[-N:] if len(fft_mag) >= N else np.pad(fft_mag, (0, N-len(fft_mag)), 'constant')
        
        # Spectral Flatness: geometric mean / arithmetic mean
        # geometric mean = exp(mean(log(fft_mag)))
        # handle zeros
        nonzero_mag = fft_mag[fft_mag > 0]
        if len(nonzero_mag) > 0:
            geom_mean = np.exp(np.mean(np.log(nonzero_mag)))
            spectral_flatness = geom_mean / (np.mean(fft_mag) + 1e-12)
        else:
            spectral_flatness = 0.0
        
        # Spectral Spread: measure variance around centroid
        # spread = sqrt( sum((f - centroid)^2 * A(f)) / sum(A(f)) )
        if np.sum(fft_mag) > 0:
            spectral_spread = np.sqrt(np.sum(((freqs - spectral_centroid)**2) * fft_mag) / np.sum(fft_mag))
        else:
            spectral_spread = 0.0
        
        # Return features in a list
        features_fft = [
            fft_mean, fft_std, fft_median, fft_range,
            dominant_freq, dominant_amp,
            low_energy, mid_energy, high_energy,
            spectral_entropy, spectral_centroid,
            spectral_flatness, spectral_spread
        ]
        # Add top N FFT coefficients
        features_fft += list(top_n_coeffs)
        
        return features_fft

    # Compute FFT features for each axis
    for arr in [x_arr, y_arr, z_arr]:
        features += compute_fft_features(arr, fs=sampling_rate)

    ########################################
    # PSD-based Features (Welch’s method)   #
    ########################################
    def compute_psd_features(arr, fs=sampling_rate):
        f_welch, Pxx = welch(arr, fs=fs, nperseg=window_size)
        # Pxx is the PSD estimate: units of V^2/Hz (assuming raw accel units^2)
        
        total_power = np.sum(Pxx)
        
        # Define frequency bands: low (0–3 Hz), mid (3–10 Hz), high (>10 Hz)
        low_idx = (f_welch >= 0) & (f_welch <= 3)
        mid_idx = (f_welch > 3) & (f_welch <= 10)
        high_idx = (f_welch > 10)
        
        low_power = np.sum(Pxx[low_idx])
        mid_power = np.sum(Pxx[mid_idx])
        high_power = np.sum(Pxx[high_idx])
        
        # Relative power in bands
        if total_power > 0:
            low_rel = low_power / total_power
            mid_rel = mid_power / total_power
            high_rel = high_power / total_power
        else:
            low_rel = mid_rel = high_rel = 0.0
        
        # Peak power and frequency
        max_idx = np.argmax(Pxx)
        peak_freq = f_welch[max_idx]
        peak_power = Pxx[max_idx]
        
        # Spectral Entropy (PSD-based)
        if total_power > 0:
            p = Pxx / total_power
            psd_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            psd_entropy = 0.0
        
        # Bandpower Ratios: e.g. Low/High
        # Avoid division by zero
        if high_power > 0:
            low_high_ratio = low_power / high_power
        else:
            low_high_ratio = 0.0
        
        # Another ratio example: low to mid
        if mid_power > 0:
            low_mid_ratio = low_power / mid_power
        else:
            low_mid_ratio = 0.0
        
        # Spectral centroid from PSD: sum(f*Pxx)/sum(Pxx)
        if total_power > 0:
            psd_centroid = np.sum(f_welch * Pxx) / total_power
        else:
            psd_centroid = 0.0
        
        # Spectral spread from PSD
        if total_power > 0:
            psd_spread = np.sqrt(np.sum((f_welch - psd_centroid)**2 * Pxx) / total_power)
        else:
            psd_spread = 0.0
        
        # Spectral flatness from PSD: geometric mean of Pxx over arithmetic mean
        nonzero_pxx = Pxx[Pxx > 0]
        if len(nonzero_pxx) > 0:
            geom_mean = np.exp(np.mean(np.log(nonzero_pxx)))
            psd_flatness = geom_mean / (np.mean(Pxx) + 1e-12)
        else:
            psd_flatness = 0.0

        return [
            total_power,
            low_rel, mid_rel, high_rel,
            peak_freq, peak_power,
            psd_entropy,
            low_high_ratio, low_mid_ratio,
            psd_centroid, psd_spread, psd_flatness
        ]
    
    # Compute PSD features for each axis
    for arr in [x_arr, y_arr, z_arr]:
        features += compute_psd_features(arr, fs=sampling_rate)

    return np.array(features)
```

```python
from scipy import stats, signal
import numpy as np
import pandas as pd
from scipy.signal import find_peaks, butter, filtfilt, welch

def feature_engineering(x_list, y_list, z_list, sampling_rate=66.7):
    # Convert lists to arrays and handle NaN
    x_arr, y_arr, z_arr = np.nan_to_num(np.array(x_list)), np.nan_to_num(np.array(y_list)), np.nan_to_num(np.array(z_list))
    window_size = len(x_arr)
    features = []
    
    # Filtering functions
    def butter_filter(data, cutoff, fs, order, filter_type):
        nyquist = 0.5 * fs
        normal_cutoff = cutoff / nyquist
        b, a = butter(order, normal_cutoff, btype=filter_type, analog=False)
        return filtfilt(b, a, data)

    # Apply low-pass and high-pass filters
    x_filt = butter_filter(x_arr, cutoff=15.0, fs=sampling_rate, order=4, filter_type='low')
    y_filt = butter_filter(y_arr, cutoff=15.0, fs=sampling_rate, order=4, filter_type='low')
    z_filt = butter_filter(z_arr, cutoff=15.0, fs=sampling_rate, order=4, filter_type='low')

    x_filt = butter_filter(x_filt, cutoff=0.2, fs=sampling_rate, order=2, filter_type='high')
    y_filt = butter_filter(y_filt, cutoff=0.2, fs=sampling_rate, order=2, filter_type='high')
    z_filt = butter_filter(z_filt, cutoff=0.2, fs=sampling_rate, order=2, filter_type='high')

    # Time-domain features
    stats_dict = {}
    for axis, arr in zip(['x', 'y', 'z'], [x_filt, y_filt, z_filt]):
        axis_mean = np.mean(arr)
        axis_std = np.std(arr)
        # Data-driven prominence for peak detection
        peak_prominence = 0.5 * axis_std if axis_std > 0 else 0.01
        peaks, _ = find_peaks(arr, prominence=peak_prominence)
        
        # Zero-crossing rate (ZCR)
        zcr = np.sum(np.diff(np.sign(arr)) != 0) / window_size
        
        stats_dict[axis] = {
            'mean': axis_mean,
            'std': axis_std,
            'median': np.median(arr),
            'min': np.min(arr),
            'max': np.max(arr),
            'mad': np.median(np.abs(arr - np.median(arr))),
            'aad': np.mean(np.abs(arr - axis_mean)),
            'peaks': len(peaks),
            'zcr': zcr
        }

    for axis in ['x', 'y', 'z']:
        s = stats_dict[axis]
        features += [s['mean'], s['std'], s['mad'], s['aad'], s['min'], s['max']]
        features.append(s['max'] - s['min'])  # range
        features.append(s['peaks'])          # peak count
        features.append(s['zcr'])            # zero-crossing rate
    
    # Resultant and SMA
    resultant = np.sqrt(x_filt**2 + y_filt**2 + z_filt**2)
    features.append(np.mean(resultant))  # Avg resultant
    sma = (np.sum(np.abs(x_filt)) + np.sum(np.abs(y_filt)) + np.sum(np.abs(z_filt))) / window_size
    features.append(sma)  # Signal Magnitude Area

    ########################################
    # Frequency-domain features (FFT-based)
    ########################################
    def compute_fft_features(arr, fs=sampling_rate):
        fft_vals = np.fft.fft(arr)
        freqs = np.fft.fftfreq(window_size, d=1.0/fs)
        
        # Positive half spectrum
        half_mask = freqs > 0
        freqs = freqs[half_mask]
        fft_mag = np.abs(fft_vals[half_mask])
        power_spectrum = fft_mag**2
        
        # Band energies
        low_band = (freqs >= 0) & (freqs <= 3)
        fog_band = (freqs >= 6) & (freqs <= 8)
        
        low_energy = np.sum(power_spectrum[low_band])
        fog_energy = np.sum(power_spectrum[fog_band])
        total_energy = np.sum(power_spectrum)
        
        # Freeze Index
        freeze_index = fog_energy / (low_energy + 1e-12)
        
        # Spectral Entropy
        if total_energy > 0:
            p = power_spectrum / total_energy
            spectral_entropy = -np.sum(p * np.log2(p + 1e-12))
        else:
            spectral_entropy = 0.0
        
        # Spectral Centroid
        spectral_centroid = np.sum(freqs * fft_mag) / np.sum(fft_mag) if np.sum(fft_mag) > 0 else 0.0

        return [low_energy, fog_energy, freeze_index, spectral_entropy, spectral_centroid]
    
    for arr in [x_filt, y_filt, z_filt]:
        features += compute_fft_features(arr, fs=sampling_rate)

    ########################################
    # Power Spectral Density (PSD-based)
    ########################################
    def compute_psd_features(arr, fs=sampling_rate):
        f_welch, Pxx = welch(arr, fs=fs, nperseg=window_size)
        
        # Frequency band indices
        low_idx = (f_welch >= 0) & (f_welch <= 3)
        fog_idx = (f_welch >= 6) & (f_welch <= 8)
        
        # Band powers
        low_power = np.sum(Pxx[low_idx])
        fog_power = np.sum(Pxx[fog_idx])
        total_power = np.sum(Pxx)
        
        # Freeze Index and Power Ratios
        freeze_index = fog_power / (low_power + 1e-12)
        low_rel = low_power / total_power if total_power > 0 else 0.0
        fog_rel = fog_power / total_power if total_power > 0 else 0.0
        
        # Spectral Centroid
        psd_centroid = np.sum(f_welch * Pxx) / total_power if total_power > 0 else 0.0
        
        return [low_rel, fog_rel, freeze_index, psd_centroid]

    for arr in [x_filt, y_filt, z_filt]:
        features += compute_psd_features(arr, fs=sampling_rate)

    # Construct column names
    axes = ['x', 'y', 'z']
    time_features = ['mean', 'std', 'mad', 'aad', 'min', 'max', 'range', 'peaks', 'zcr']
    # 3 axes * 9 features each = 27, plus 2 global = 29
    columns = []
    for axis in axes:
        for feat in time_features:
            columns.append(f"{axis}_{feat}")
    columns += ['mean_resultant', 'sma']

    # FFT-based features per axis: 5 per axis, total 15
    fft_feats = ['fft_low_energy', 'fft_fog_energy', 'fft_freeze_index', 'fft_spectral_entropy', 'fft_spectral_centroid']
    for axis in axes:
        for feat in fft_feats:
            columns.append(f"{axis}_{feat}")

    # PSD-based features per axis: 4 per axis, total 12
    psd_feats = ['psd_low_rel', 'psd_fog_rel', 'psd_freeze_index', 'psd_centroid']
    for axis in axes:
        for feat in psd_feats:
            columns.append(f"{axis}_{feat}")

    # Create a DataFrame from the features
    df = pd.DataFrame([features], columns=columns)

    return df
```
